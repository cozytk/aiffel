````python
# 필요한 모델 import
from sklearn.datasets import load_digits, load_wine, load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
````

````python
# 데이터 준비
digits = load_digits()
wines = load_wine()
breast_cancers = load_breast_cancer()
````

````python
# 다양한 모델로 학습시켜보기
'''
Decision Tree 사용해 보기
Random Forest 사용해 보기
SVM 사용해 보기
SGD Classifier 사용해 보기
Logistic Regression 사용해 보기
'''
````

````python
# logistic_model에서 solver를 안넣어주니 iteration이 초과하는 오류가 발생해서 넣어주었습니다

from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(random_state=42)

from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(random_state=42)

from sklearn import svm
svm_model = svm.SVC(random_state=42)

from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier(random_state=42)

from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(random_state=42, solver='liblinear')
````

````python
# 평가 지표로는 classification_report 사용. 그냥 이거저거 다 들어있길래..

def compare_models(data):
  # 데이터 이해하기
  dir(data)
  data_data = data.data
  data_label = data.target
  print(data.target_names)
  print(data.DESCR)

  # train test 분리
  X_train, X_test, y_train, y_test = train_test_split(data_data, data_label, test_size=0.2, random_state=42 )
  models = [decision_tree, random_forest, svm_model, sgd_model, logistic_model]
  for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred))
````

````python
# 손글씨 데이터

compare_models(digits)
````

출력결과 세번째에 해당하는 svm_model이 결과가 가장 좋은 것 같습니다!

손글씨 분류자체는 스펨메일이나 암판별 문제와는 다르게 class가 10개여서 class가 2개인 데이터로만 recall과 precision 중에 어떤게 나을지 고민해봐서
10개의 클래스를 가진 digits 데이터는 어떻게 적용할지는 모르겠습니다.
밑에 코드를 보아하니 데이터도 골고루 들어있는 것 같으니 accuracy가 무난히 좋을 것 같습니다.
따라서 accuracy가 가장 좋은 평가지표라고 생각했습니다.
